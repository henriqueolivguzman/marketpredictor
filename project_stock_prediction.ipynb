```python
import yfinance as yf;
import pandas as pd

sp500 = yf.Ticker("^GSPC"); #classe .Ticker()baixa os preços das ações, no caso estamos baixando os preços do S&P

sp500 = sp500.history(period="max")
```

```python
sp500 #é retornado um datframe com os dados a respeito dos preços
```

```python
sp500.plot.line(y="Close", use_index = True)#sera plotado grafico que no eixo y tera os preços de fechamento e x os index que no caso são as data
```

```python
#iremos apagar as colunas desnecessarias, dividendos e stocksplits

del sp500["Dividends"]
del sp500["Stock Splits"]
```

```python
#quremos um lgoritmo que traga não valores absolutos referentes a preço mas sim referente a se vai subir ou descer

#vamos criar uma coluna chamada tomorrow que ira comparar apresentar o preço de abertura e fechamento do dia seguinte

sp500["tomorrow"] = sp500["Close"].shift(-1) 

sp500
```

```python
'''Agora sera criada uma coluna target que basicamente ira compara baseado no preço de hoje se amanha o preço
de hoje se amanha ira ou não subir o preço. isto retornara um booleano porem faremos uma conversão para inteiro
AS REFERENCIAS CONSIDERADAS SÃO OS PREÇOS DE FECHAMENTO
''' 

sp500["target"] = (sp500["tomorrow"]> sp500["Close"]).astype(int)
sp500
```

```python
'''uma vez que o preço subiu temos target igual a 1 e quando não target igual a zero. Fica claro aqui que no final iremos
chegar em algo como um classificador binario contudo seguiremos a implementação do algoritmo para ter certeza
'''
```

```python
'''a partir de agora sabemos que não queremos todos os dados de preços pois os mais antigos acabam sendo irrelevantes, de 1928, 
1929 portanto utilizaremos um "filtro", este filtro sera criado a partir dos dados de 1999 para frente, isto significa que consideraremos
dados somente a partir de 1999
'''
sp500 = sp500.loc["1990-01-01": ].copy() #copy utilizado para não receber warning

sp500
```

```python
'''sera utilizado um classificador chamado random forest'''

from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 100, min_samples_split = 100, random_state = 1)
'''teremos 100 arvores de decisão que serao treinadas
min samples split ira proteger do overfitting
ramdomstate = 1 significa que se fizermos a implementação com os argumentos definidos o modelo produzira valores iguais 
independente do numero de vezes que seja feita a inicialização do modelo 
'''

#selecionando os dados de treinamento e teste
train = sp500.iloc[:-100] #todos os dados exceto os ultimos 100
test = sp500.iloc[-100:]#todos os 100 ultimos dados 


```

```python
#determinando os elementos que serão considerados nas predições

predictors =["Close", "Volume", "High", "Low"]

#A partir dos elementos tentaremos prever a coluna target 
model.fit(train[predictors], train["target"])
```

```python
#a partir da inicialização do treinamento do modelo iremos analisar o quão bom/preciso esta este modelo
import pandas as pd
from sklearn.metrics import precision_score

'''a precisão ira definir se quando a coluna target fo modelo resultou em 1 o rótulo desta era o mesmo, se sim o modelo esta 
preciso, qual é a porcentagem de acertos destes um em relação ao que realmente aconteceu'''

#uma vez treinado o modelo vamos testa-lo com os dados de teste
preds = model.predict(test[predictors])

preds = pd.Series(preds, index=test.index) #o array preds sera convertido em uma serie de pandas

#agora iremos determinar a precisão uma vez que sabemos os rotulos de test, ultimos 100 dados, iremos compara-los
precision_score(test["target"], preds)
```

```python
'''a precisão foi de 50%, ou seja metade do tempo que foi predito que as ações subiriam, no grupo de teste,
o modelo acertou contudo é importante pensar se é um bom numero ou preferimos algo mais confiavel.. vamos melhorar
'''

#iremos plotar os dados dos rotulos e os preditos para visualizar a situação

combined = pd.concat([test["target"], preds], axis= 1)#dados serão apresentados juntos no grafico e os valores ficarão no eixo y

combined.plot()
```

```python
"""é necessario criar uma estrutura de testes do modelo melhor do que a utilizada, poderia ser feita através de 
testes que utilizem conjuntos especificos de anos do sp, anos em que houveram situações especificas que poderiam  ser consideradas
Para tal sera feito algo chamado backtesting, utilizado para determinar como o modelo se sai utilizando periodos historicos
"""
#sera criada uma função para isso
def predict(train, test, predictors, model):
    model.fit(train[predictors], train["target"]) #a partir das features definidas em predictors queremos "acertar" o target
    preds = model.predict(test[predictors])
    preds = pd.Series(preds, index = test.index , name = "Predictions")
    combined = pd.concat([test["target"], preds], axis = 1)
    return combined

#Sera definida a funcção backtest
'''explicando o argumento start: ano tem, para o mercado, 250 dias portanto o primeiro modelo sera treinado com 2500 dias
ou dez anos de dados, estes serão utilizados para teste do 11º ano, após isso temos o argumento step, este fala sobre como
após os dez primeiros anos o algoritmo ira progredir no treinamento no caso este ira considerar os 11 primeiros anos para treinar
os 12ºano e depois sera treinado os 12 anos para o modelo ser testado no 13º ano e assim por 
'''
def backtest(data, model, predictors, start = 2500, step= 250):
    all_predictions = [] #lista que contem em cada posição a predição na forma dataframe para um ano que sera o teste 
    
    for i in range(start, data.shape[0], step):
        train = data.iloc[0:i].copy()
        test = data.iloc[i: (i+step)].copy()
        predictions = predict(train, test , predictors, model)
        all_predictions.append(predictions)
    return pd.concat(all_predictions)

predictions = backtest(sp500, model, predictors)
```

```python
'''Agora faremos a contagem das predições que resultaram em um e zero'''
predictions["Predictions"].value_counts()
```

```python
#determinando a precisão das predições
precision_score(predictions["target"], predictions["Predictions"])
```

```python
#adicionando mais preditores ao modelo, ou seja mais elementos a serem consideradas
"""serão criadas algumas medias moveis, e portanto serão criadas razões dos preços atuais com preços comparados a
uma semana atras, um mes atras, um ano atras etc, estamos , a grosso modo, criando mais elementos de comparacao00000000000000000000000000000...
"""
horizons = [2, 5, 60, 250, 1000];#sera feita comparação do preço atual com preços passados 

new_predictors = []

for horizon in horizons:
    rolling_averages = sp500.rolling(horizon).mean() #determinação da media movel
    
    ratio_column = f"Close_Ratio_{horizon}" #novo preditor 1
    
    sp500[ratio_column] = sp500["Close"]/rolling_averages["Close"]
    
    trend_column = f"Trend_{horizon}" #numero de dias que as ações efetivamente subiram , novo preditor 2
    sp500[trend_column] = sp500.shift(1).rolling(horizon).sum()["target"] 
    '''o código acima ira tomar o periodo anterior, definido no horizons
    e ira fazer uma soma dos targets no periodo, por exemplo de 5 dias, nestes cinco dias anteriores quantos dias 
    houve efetivamente elevação no preço'''
    
    new_predictors += [ratio_column, trend_column]
    
    
    
```
